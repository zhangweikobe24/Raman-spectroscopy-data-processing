```python
# score plot of PCA-LDA
from numpy import mean
import pandas as pd
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# get the dataset
data = pd.read_csv('macrophage.csv')
X = data.iloc[:, 1:1301]
y = data.iloc[:,0]


pca = PCA(n_components=2)
pca.fit(X)
result = pca.transform(X)
components = pd.DataFrame(pca.components_, columns = X.columns, index=[1, 2])
components = components.T
components.columns = ['Principle_Component_1', 'Principle_Component_2']

lda = LinearDiscriminantAnalysis()
canonicals = lda.fit(result, y).transform(result)
canonicalsDf = pd.DataFrame(data = canonicals, columns = ['LDA 1', 'LDA 2'])

data[['group']].head()

finalDf = pd.concat([canonicalsDf, data[['group']]], axis = 1)
finalDf.head(712)

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot (1,1,1)
ax.set_xlabel('Canonical Variable 1', fontsize = 15)
ax.set_ylabel('Canonical Variable 2', fontsize = 15)
ax.set_title('PCA-LDA_Macrophage', fontsize = 20)

targets = ['M0', 'M1', 'M2']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['group'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'LDA 1']
               , finalDf.loc[indicesToKeep, 'LDA 2']
               , c = color
               , s = 30)
    
    ax.legend(targets)
    ax.grid()

plt.savefig('PCA-LDA_Macrophage 2 PCs.png')
```
